##Heandling missing values

```python
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

droping the null values


```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

#### Changes the required columns as integer data type

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns to float data datatype

> Indented block




```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!="Classes":
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## Let ave the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

##  Exploratory Data Analysis

```python
## drop day,month and year
df_copy=df.drop(['day','month','year'],axis=1)
```

# Plot density plots for all numerical features

```python
sns.set_style("darkgrid")  # Set Seaborn style

num_features = len(df_copy.select_dtypes(include=['number']).columns)
rows = int(np.ceil(num_features / 3))  # Adjust rows dynamically

df_copy.plot(kind='density', subplots=True, layout=(rows, 3), figsize=(20, 5 * rows), sharex=False)
plt.show()
```

```python
df_copy.hist(bins=50, figsize=(20, 15), color='skyblue')
plt.show()
```

```python
## Percentage for Pie Chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting piechart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%',colors=["orange","green"])
plt.title("Pie Chart of Classes")
plt.show()
```
